```python
import matplotlib.pyplot as plt
import csv
from matplotlib.pyplot import figure
```

```python
n = 0
training_loss = []
test_loss = []
with open('2_7_23/leaky_relu_model2.3.2.csv', 'r') as file: #reading in file
    reader = csv.reader(file)
    header = next(reader)
    for row in reader:
        if n > 100:
            break
        training_loss.append(float(row[0]))
        test_loss.append(float(row[1]))
        n+=1
```
